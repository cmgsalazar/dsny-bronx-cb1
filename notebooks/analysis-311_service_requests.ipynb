### Automating this to create dfs per year

`df_dict` will be all complaints filtered by year, while `bx1_dict` will be value counts of complaint types per year.

```python
def filter_by_year(year):
    # filter by year
    df_year = df[df["Created Date"].str.contains(str(year)) == True]
    return df_year

def count_complaints(df_year, year):
    # count complaints by year
    bx1_year = df_year["Complaint Type"].value_counts().reset_index()
    bx1_year.columns = ["Complaint Type", str(year)]
    return bx1_year
    
years_max = 2024 # latest year on df
years_min = 2014 # earliest year on df

df_dict = {}
bx1_dict = {}
errors_list = []

for year in range(years_min, years_max +1):  # +1 to include years_max
    try:
        df_dict[year] = filter_by_year(year)
        bx1_dict[year] = count_complaints(df_dict[year], year)
    except Exception as error:
        print(f"Something is wrong with {year} due to {error}.")
        errors_list.append(year, error)
    finally:
        print(f"Complaints for {year}:")
        print(bx1_dict[year])
```

```python
len(df_dict)
```

### Merging all complaint counts by year into one df

```python
df_complaint_counts = pd.concat(bx1_dict.values(), ignore_index=True)

# this cleans the df by compiling by "Complaint Type"
df_complaint_counts = df_complaint_counts.groupby("Complaint Type").agg(lambda x: x.sum()).reset_index()
df_complaint_counts
```

```python
df_complaint_counts.to_csv("parsed-data/311-complaints-by-year.csv", encoding="UTF-8", index=False)
```

### Visualizing data

```python
# choosing only first 10
df_top_10 = df_complaint_counts.sort_values(by="2024", ascending=False).head(10).reset_index(drop=True)
df_top_10
```

```python
# melting df_complaint_counts into a long format that seaborn reads
df_long = df_top_10.melt(id_vars="Complaint Type", var_name="Year", value_name="Count")
df_long
```

```python
# replace NaN with 0
df_long = df_long.fillna(0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# seaborn tutorial: https://plainenglish.io/blog/ridge-plots-with-pythons-seaborn-4de5725881af

sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0), 'axes.linewidth':2})
palette = sns.color_palette("Set2", 12)
g = sns.FacetGrid(df_long, palette=palette, row="Complaint Type", hue="Complaint Type", aspect=9, height=1.2)
g.map_dataframe(sns.kdeplot, x="Count", fill=True, alpha=1)
g.map_dataframe(sns.kdeplot, x="Count", color='black')
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, color='black', fontsize=10,
            ha="left", va="center", transform=ax.transAxes)

g.map(label, "Complaint Type")
g.fig.subplots_adjust(hspace=-.5)
g.set_titles("")
g.set(yticks=[], xlabel="Count")
g.despine( left=True)
plt.suptitle("311 Service Requests to DSNY from Bronx Community District 1", y=0.98)
```
