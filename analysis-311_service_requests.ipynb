### Merging data

```python
dfs = [bx1_2020, bx1_2021, bx1_2022, bx1_2023, bx1_2024]
df_all = dfs[0]

for df in dfs[1:]:
    df_all = pd.merge(df_all, df, on="Complaint Type", how="outer")
```
